```python
from ultralytics import YOLO
import zipfile
import os
```

```python
with zipfile.ZipFile("electronics-components.zip", 'r') as zip_ref:
    zip_ref.extractall("electronics_components")
```

```python
model = YOLO('yolov8n.pt')  # or 'yolov8s.pt' for better accuracy

results = model.train(
    data="electronics_components/data.yaml",
    epochs=60,
    imgsz=800,
    batch=8
)

```
